## Estudo sobre Pandas e suas  principais funções

```python
import pandas as pd
```

#### Criando Pandas DataFrame

```python
# Criando dataframes
df = pd.DataFrame(columns=['estado','sigla','populacao','capital'])
print(f"O dataframe possui {df.shape[0]} linhas e {df.shape[1]} colunas \n")

dict = {'estado': [], 'sigla': [], 'populacao': [], 'capital':[]}


display(df)
```

### Adicionando dados a um dataFrame

```python
dict['estado'].append('Mato Grosso')
dict['sigla'].append('MT')
dict['populacao'].append('4 Mi')
dict['capital'].append('Cuiabá')

dict['estado'].append('Goiás')
dict['sigla'].append('GO')
dict['populacao'].append('6 Mi')
dict['capital'].append('Goiânia')

dict['estado'].append('Mato Grosso')
dict['sigla'].append('MT')
dict['populacao'].append('2 Mi')
dict['capital'].append('Várzea Grande')


df_temp = pd.DataFrame(dict)
df = pd.concat([df,df_temp], ignore_index=True)

display(df)
```

### Filtrando por dados

```python
print(f"O 'display' abaixo mostra uma linha buscada a partir de um filtro  na coluna 'sigla' ")
brasil = df.loc[df['sigla'] == 'MT'] # Busca todos os dados que tem o Brasil na coluna 'estado'
display(brasil)

# Semelhante ao filtro 'WHERE' do SQL
```

### Filtrando por dados, mas também filtrando outras colunas

```python
print(f"O 'display' abaixo mostra um dataframe no qual filtra a capital por 'Cuiabá' mas tembém trás as colunas 'sigla' e 'populacao'")
capital = df.loc[df['capital']== 'Cuiabá', ['capital','sigla', 'populacao']]
display(capital)
```

## Criando um novo DataFrame


```python
itens = {
    "nome": ["Matheus", "Caio", "Luan", "Thales", "Pedro", "Emmanuel"],
    "quantidade": [20, 40, 80, 2, 7, 3],
    "valor": [30, 20, 10, 80, 100, 10],
    "item":["short", "luva", "meia", "camisa", "calça", "bola"]
}
vendas_df = pd.DataFrame(itens)
display(vendas_df)
```

--------------------------------------------------------------------------

# Adicionando Colunas

### Criando uma nova coluna a partir de outras colunas


```python
# A partir de uma coluna que existe
print("Este 'display' mostra a criação de uma nova tabela 'valor total' a partir da multiplicação de outras duas")
vendas_df['valor total'] = vendas_df['valor'] * vendas_df['quantidade']
display(vendas_df)
```

### Criando uma nova coluna com valores padrão

```python
# Criando uma nova coluna com valor padrão

print("Este 'display' mostra a criação de uma nova tabela 'imposto' distribuindo um valor padrão para todas as linhas")
vendas_df.loc[:, "imposto"] = 0

display(vendas_df)
```

---------------------------------------------------------------------------------
# Inserindo novos dados no DataFrame

```python
# Inserindo novos dados (linhas) no dataframe

# Não se usa mais o método .append()

print("Este 'display' mostra a junção de duas tabelas,'df' com a tabela 'vendas_df'")


 # concatenando vendas_df e df verticalmente (linhas) axis = 0 -> é o eixo da linha

print("Vertical")
vertical_concat = pd.concat([vendas_df, df], axis = 0, ignore_index= True)
display(vertical_concat)
```

### Adicionando Colunas 

```python
 # concatenando vendas_df e df horizontalmente (colunas) axis = 1 -> é o eixo da coluna

print("Horizontal")
horizontal_concat = pd.concat([vendas_df, df], axis=1)
display(horizontal_concat)
```

### Juntando dois DataFrames

```python
print("Juntando as duas juntamente em um único DataFrame.") # Fica igual  a vertical
all_concat = pd.concat([vendas_df, df])
display(all_concat)
```

### Excluindo linhas de um DataFrame

```python
# Excluindo linhas e Colunas


print("Dataframe Normal")
display(vendas_df)

print("Excluindo a primeira linha (index '0') no eixo (axis=0) do dataframe")

vendas_df = vendas_df.drop(0, axis=0)

display(vendas_df)
```

### Excluindo colunas de um DataFrame

```python
print("Dataframe Normal")
display(vendas_df)


print("Excluindo a coluna 'imposto' no eixo (axis=1) tabela no dataframe ")
vendas_df = vendas_df.drop("imposto", axis=1)

display(vendas_df)
```

# Tratamento para valores vazios ou 'NAN'

```python
print("tabela  original")

display(all_concat)
```

### Eliminar colunas

```python
all_concat = all_concat.dropna(how='all', axis=1) 
# Eliminar colunas que estão totalmente vazias ('NaN')

print("tabela tratada")
display(all_concat)
```

### Preencher valores vazios com a média da coluna

```python
# fillna -> preenche as linhas vazias da coluna selecionada com a média  dessa coluna
vendas_df.loc[:, "imposto"] = 0
print("tabela original")
display(vendas_df)

vendas_df['imposto'] = vendas_df['imposto'].fillna(vendas_df['imposto'].mean()) 

print("tabela com a media")

display(vendas_df)
```

### Deletando linhas com pelo menos 1 valor vazio

```python
# Deletando linhas que possuem pelo menos 1 valor vazio
print("Deletando linhas com pelo menos 1 valor vazio")
all_concat = all_concat.dropna() 

display(all_concat)

# Deletou todas, pois todas as linhas possuem pelo menos uma linha vazia
```

## Calcular Indicadores


#### Com .value_counts()

```python
novas_vendas = {
    "nome": ["Matheus", "Caio", "Luan", "Thales", "Pedro", "Emmanuel", "Pedro"],
    "quantidade": [20, 40, 80, 2, 7, 3, 2],
    "valor": [30, 20, 10, 80, 100, 10, 10],
    "item":["short", "luva", "meia", "camisa", "calça", "bola", "bola"],
    "ID Loja": ["Iguatemi Esplanada", "Norte Shopping", "Center Shopping", "Ribeirão Shopping", "Norte Shopping", "Rio Mar Shopping Fortaleza", "Center Shopping"]
}
novas_vendas = pd.DataFrame(novas_vendas)

print("Tabela Original")
display(novas_vendas)


# Value Counts

transacoes_loja =novas_vendas['ID Loja'].value_counts()
display(transacoes_loja)
```

#### Com groupby()


```python
faturamento_produto = novas_vendas[['item', 'valor']].groupby('item').sum()

display(faturamento_produto)
```
